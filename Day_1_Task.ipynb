```python

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer


nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
def preprocess_text(passage):
    # Tokenize the passage
    words = word_tokenize(passage.lower())  # Convert to lowercase


    stop_words = set(stopwords.words('english'))


    filtered_words = [word for word in words if word.isalnum() and word not in stop_words]

    # Stemming
    stemmer = PorterStemmer()
    stemmed_words = [stemmer.stem(word) for word in filtered_words]

    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    lemmatized_words = [lemmatizer.lemmatize(word) for word in stemmed_words]


    return ' '.join(lemmatized_words)


passage = """
Data analytics: AI can analyze large amounts of data to identify patterns and relationships.
Natural language processing: AI can understand and respond to human language.
Speech recognition: AI can understand human speech by breaking it down into sounds and analyzing them.
Machine vision: AI can see and interpret images.
Virtual assistants: AI can help users with tasks like scheduling, making calls, and providing directions.
Ethical concerns
AI can raise ethical concerns about privacy, employment, and socioeconomic inequality.
AI can also raise concerns about job displacement, which can be addressed by retraining and upskilling the workforce.
"""


preprocessed_passage = preprocess_text(passage)
print("Original Passage:")
print(passage)
print("\nPreprocessed Passage:")
print(preprocessed_passage)
```
